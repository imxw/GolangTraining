## helloworld

```go
package main

// fmt 包实现了格式化IO函数
import "fmt"

func main(){
    
    fmt.Println("1: hello, world!")
    // %d 十进制占位符
    // %b 二进制占位符
    fmt.Printf("2: %d - %b \n", 42,42)
    // %x 十六进制占位符，字母形式为小写 a-f
    fmt.Printf("3: %d - %b - %x \n", 42,42,42)
    // # 为八进制添加前导 O，为十六进制添加前导 Ox, 为%p(%#p) 去掉前导 Ox，如果是可打印字符，%U（%#U）会写出该字符的 Unicode 编码形式
    fmt.Printf("4: %d - %b - %#x \n", 42, 42, 42)
    fmt.Printf("5: %d - %b - %#X \n", 42, 42, 42)    
    fmt.Printf("6: %d \t %b \t %#X \n", 42, 42, 42)
}
main()
```

## 循环

```go
package main

import "fmt"

func main() {
    for i := 1000000; i < 1000100; i++ {
        fmt.Printf("%d \t %b \t %x \n", i, i, i)
    }
}
main()
```

```go
package main

import "fmt"

func main() {
    for i := 60; i < 122; i++ {
        fmt.Printf("%d \t %b \t %x \t %q \n", i, i, i, i)
    }
}
main()
```
